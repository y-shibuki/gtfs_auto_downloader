```python
import time
from collections import defaultdict

def crawlers_1():
    print("1秒ごとの処理を実行しました")
    time.sleep(0.5)

def crawlers_2():
    print("2秒ごとの処理を実行しました")

def crawlers_3():
    print("3秒ごとの処理を実行しました")

intervals = {
    crawlers_1: 1,
    crawlers_2: 2,
    crawlers_3: 3
}
```

```python

last_execution = {}
time_before = {}
res_time = defaultdict(list)
N = 60
```

```python
f = True
while f:
    current_time = time.time()

    for func, interval in intervals.items():
        if func not in last_execution or current_time - last_execution[func] >= interval:
            last_execution[func] = current_time
            func()
            
            now = time.time()
            if func in time_before:
                res_time[func].append(now - time_before[func])
            time_before[func] = now
            f = (len(res_time[func]) < N)

    processing_time = interval = time.time() - current_time
    time.sleep(1 - processing_time)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(res_time[crawlers_1], bins=30, alpha=0.5)

# グラフの表示
plt.show()
```
